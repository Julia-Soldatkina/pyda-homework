## Задание 1

Напишите функцию, которая принимает на вход строку и проверяет является ли она валидным транспортным номером (1 буква, 3 цифры, 2 буквы, 2-3 цифры). Обратите внимание, что не все буквы кириллического алфавита используются в транспортных номерах.

Если номер валиден, то функция должна возвращать отдельно номер и регион.

Примеры работы программы:  

1. `car_id = 'A222BC96'`

Результат: 
`Номер A222BС валиден. Регион: 96`

2. `car_id = 'АБ22ВВ193'`

Результат: 
`Номер не валиден`

````python
import re
````

````python
def check_car_id():
    car_id = input('Введите номер транспортного средства')
    if re.match(r'[АВЕКМНОРСТУХ]{1}[0-9]{3}[АВЕКМНОРСТУХ]{2}[0-9]{2,3}',car_id):
        print(f' Номер {car_id} валиден. Регион: {car_id[-2:]}')
    else:
        print('Номер не валиден')
    return
````

````python
check_car_id()
````

## Задание 4

Напишите функцию, которая будет принимать на вход список email-адресов и выводить их распределение по доменным зонам.

Пример работы программы:  

`emails = ['test@gmail.com', 'xyz@test.in', 'test@ya.ru', 'xyz@mail.ru', 'xyz@ya.ru', 'xyz@gmail.com']`

Результат: 
```
gmail.com: 2
test.in: 1 
ya.ru: 2 
mail.ru: 1
```

````python
emails = ['test@gmail.com', 'xyz@test.in', 'test@ya.ru', 'xyz@mail.ru', 'xyz@ya.ru', 'xyz@gmail.com']

def count_domen():
    res ={}
    count= 1
    email_pattern = r'@\w+.\w+'

    for el in emails:
        d = '\t'.join(re.findall(email_pattern,el))
        if d in res:
            res[d] += count
        else:
            res[d] = count
        
    for k,v in res.items():
        print(f' {k}:{v}')
````

````python
count_domen()
````


````python
some_text = 'Эталонной реализацией Python является интерпретатор CPython, поддерживающий большинство активно используемых платформ. Он распространяется под свободной лицензией Python Software Foundation License, позволяющей использовать его без ограничений в любых приложениях, включая проприетарные.'
````

**Вариант 1**

````python
def word():
    vowel = re.findall(r'\b[аеиоуяюэАЕИОУЯЮЭaeiouAEIOU]\w+',some_text)
    consonant = re.findall(r'\b[^аеиоуяюэАЕИОУЯЮЭaeiouAEIOU ]\w',some_text)
    print(f' Слов на гласные буквы:{len(vowel)}')
    print(f' Слов на согласные буквы:{len(consonant)}')
````

````python
word()
````

*В данном случае не учитывается предлог "в" при подсчете слов на согласную. Затрудняюсь в исправлении регулярного выражения. Как альтернатива, решение может быть таким*
